```python
## Random the posistion of where to go (x,y)
```

```python
def metropolis(N, dr, data, size):

    x_chain = np.zeros(N)
    y_chain = np.zeros(N)
    P_chain = np.zeros(N)
    P_chain[1] = 1

    for i in range(1,N):
        theta = random.random(1)*2*np.pi
        dx = cos(theta)
        dy = sin(theta)
        x_new = x_chain[i-1] + dx*dr
        y_new = y_chain[i-1] + dy*dr
        if x_new>size:
            x_new=x_new-size
        if y_new>size:
            y_new=y_new-size  
        if x_new<0:
            x_new=size+x_new
        if y_new<0:
            y_new=size+y_new

        P_chain[i] = data[int(x_new), int(y_new)]
        if P_chain[i]/P_chain[i-1]>=1:
            x_chain[i]=x_new
            y_chain[i]=y_new
        else:
            alpha = random.random(1)
            if alpha>0.5:
                x_chain[i]=x_new
                y_chain[i]=y_new
            else:
                x_chain[i]=x_chain[i-1]
                y_chain[i]=y_chain[i-1]
    return x_chain, y_chain
```

```python
x1, y1 = metropolis(100000, 3, data_set1, 64)
figure(figsize=(6,6))
scatter(x1, y1, alpha=0.01, s=40, c='k')
```

```python
x1, y1 = metropolis(100000, 3, data_set2, 64)
figure(figsize=(6,6))
scatter(x1, y1, alpha=0.01, s=40, c='k')
```
